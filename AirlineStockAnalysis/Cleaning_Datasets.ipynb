```python
#import dependencies
import pandas as pd
from datetime import datetime
```

```python
#read american airlines csv
american_df = pd.read_csv("data/5yrAmericanAirlines.csv")

#display dataframe
american_df.head()
```

```python
#determining column names
american_df.columns
```

```python
#renaming columns to differentiate between datasets
american_air = american_df.rename(columns={" Close/Last":"AAL_Close"," Volume":"AAL_Volume"," Open":"AAL_Open"," High":"AAL_High"," Low":"AAL_Low"})

#display dataframe
american_air.head()
```

```python
#read jetblue airlines csv
jetblue_df = pd.read_csv("data/5yrJetBlue.csv")

#display dataframe
jetblue_df.head()
```

```python
#renaming columns to differentiate between datasets
jetblue = jetblue_df.rename(columns={" Close/Last":"JBLU_Close"," Volume":"JBLU_Volume"," Open":"JBLU_Open"," High":"JBLU_High"," Low":"JBLU_Low"})

#display dataframe
jetblue.head()
```

```python
#read southwest airlines csv
southwest_df = pd.read_csv("data/5yrSouthwest.csv")

#display dataframe
southwest_df.head()
```

```python
#renaming columns to differentiate between datasets
southwest = southwest_df.rename(columns={" Close/Last":"LUV_Close"," Volume":"LUV_Volume"," Open":"LUV_Open"," High":"LUV_High"," Low":"LUV_Low"})

#display dataframe
southwest.head()
```

```python
#read spirit airlines csv
spirit_df = pd.read_csv("data/5yrSpiritAirlines.csv")

#display dataframe
spirit_df.head()
```

```python
#renaming columns to differentiate between datasets
spirit = spirit_df.rename(columns={" Close/Last":"SAVE_Close"," Volume":"SAVE_Volume"," Open":"SAVE_Open"," High":"SAVE_High"," Low":"SAVE_Low"})

#display dataframe
spirit.head()
```

```python
#read united airlines csv
united_df = pd.read_csv("data/5yrUnitedAirlines.csv")

#display dataframe
united_df.head()
```

```python
#renaming columns to differentiate between datasets
united_air = united_df.rename(columns={" Close/Last":"UAL_Close"," Volume":"UAL_Volume"," Open":"UAL_Open"," High":"UAL_High"," Low":"UAL_Low"})

#display dataframe
united_air.head()
```

```python
#read delta airlines csv
delta_df = pd.read_csv("data/5yrDeltaAirlines.csv")

#display dataframe
delta_df.head()
```

```python
#renaming columns to differentiate between datasets
delta_air = delta_df.rename(columns={" Close/Last":"DAL_Close"," Volume":"DAL_Volume"," Open":"DAL_Open"," High":"DAL_High"," Low":"DAL_Low"})

#display dataframe
delta_air.head()
```

```python
#starting the merging of dataframes
df1 = pd.merge(american_air, jetblue, how='outer', on='Date')
df1.head()
```

```python
#merging dataframes
df2 = pd.merge(southwest, spirit, how='outer', on='Date')
df2.head()
```

```python
#merging dataframes
df3 = pd.merge(united_air, delta_air, how='outer', on='Date')
df3.head()
```

```python
#merging dataframes
df4 = pd.merge(df1, df2, how='outer', on='Date')
df4.head()
```

```python
#final merged dataframe
df5 = pd.merge(df3, df4, how='outer', on='Date')
df5.head()
```

```python
#checking columns of final merged dataframe to ensure all dataframes are included
df5.columns
```

```python
#creating a copy of the final merged dataframe to work with
airlines_df = df5.copy()
airlines_df.head()
```

```python
#converting the date column from object to datetime, in order to better manipulate the data for graphs
airlines_df['Date'] = pd.to_datetime(airlines_df['Date'])
airlines_df.head()
```

```python
#pulling out the day, month, and year from the date column into their own seperate columns
airlines_df['Year'] = airlines_df['Date'].dt.year
airlines_df['Month'] = airlines_df['Date'].dt.month
airlines_df['Day'] = airlines_df['Date'].dt.day
```

```python
#pulling just the 2018-2020 data
airlines_df = airlines_df[(airlines_df.Year >= 2018) & (airlines_df.Year <= 2020)]
airlines_df
```

```python
#making a copy of the complete dataset
all_data_df = airlines_df.copy()
all_data_df
```
